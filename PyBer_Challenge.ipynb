# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
city_data_df
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
total_rides_per_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_per_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_per_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_per_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_amount_fares_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_amount_fares_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride_city_type = total_amount_fares_city_type / total_rides_per_city_type
avg_fare_ride_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver_city_type = total_amount_fares_city_type / total_drivers_per_city_type
avg_fare_driver_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides": total_rides_per_city_type, "Total Drivers": total_drivers_per_city_type, "Total Fares": total_amount_fares_city_type, "Average Ride Fare": total_amount_fares_city_type, "Average Driver Fare": avg_fare_driver_city_type})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,.2f}".format)

pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)

pyber_summary_df["Average Ride Fare"] = pyber_summary_df["Average Ride Fare"].map("${:,.2f}".format)

pyber_summary_df["Average Driver Fare"] = pyber_summary_df["Average Driver Fare"].map("${:,.2f}".format)


pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_amount_fares_per_city_type_and_date = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
total_amount_fares_per_city_type_and_date
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_amount_fares_per_city_type_and_date = total_amount_fares_per_city_type_and_date.reset_index()
total_amount_fares_per_city_type_and_date
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_amount_fares_per_city_type_and_date = total_amount_fares_per_city_type_and_date.pivot(index= 'date', columns = 'type', values ='fare')
total_amount_fares_per_city_type_and_date
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
From_Date1_to_Date2_df = total_amount_fares_per_city_type_and_date.loc["2019-01-01":"2019-04-29"]
From_Date1_to_Date2_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
From_Date1_to_Date2_df.index = pd.to_datetime(From_Date1_to_Date2_df.index)
From_Date1_to_Date2_df
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
From_Date1_to_Date2_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_by_week_df = From_Date1_to_Date2_df.resample('W').sum()
fares_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fares_by_week_df.plot(figsize=(12,4))

plt.title("Total Fare by Type of City", fontsize = 18)
plt.ylabel("Fare ($)", fontsize = 14)
plt.xlabel("Month", fontsize = 14)
plt.grid(True)

plt.legend(bbox_to_anchor = (1,1), loc = 'upper left')
plt.tight_layout()

plt.show()
plt.savefig("analysis/PyBer_fare_summary.png")
```
